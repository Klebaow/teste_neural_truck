```python
import random
import pandas as pd

num_obs = 10000

velocidade_atual = []
aceleração_caminhão = []
posi_acelerador = []
posi_freio = []
rota_motor = []

for _ in range(num_obs):
    acelerador = random.randint(0, 100)  # %
    freio = random.randint(0, 100)  # %
    velocidade = max(0, acelerador - freio)  # km/h
    aceleração = max(0, (acelerador - freio) / 10)  # m/s^2
    rotação = acelerador * 30  # RPM, assumindo 3000 RPM no máximo quando o acelerador está em 100%

    velocidade_atual.append(velocidade)
    aceleração_caminhão.append(aceleração)
    posi_acelerador.append(acelerador)
    posi_freio.append(freio)
    rota_motor.append(rotação)

carga_truck = [random.randint(0, 30000) for _ in range(num_obs)]  # kg
cond_estrada = [random.choice([0, 1, 2]) for _ in range(num_obs)]  # Ruim, Média, Boa

consu_combus = [5 + carga/6000 for carga in carga_truck]  # litros por 100km

cond_condu = [0 if cond == 0 else random.choice([0, 1]) for cond in cond_estrada]  # Insegura, Segura

data = {
    "velocidade_atual": velocidade_atual,
    "aceleração_caminhão": aceleração_caminhão,
    "posi_acelerador": posi_acelerador,
    "posi_freio": posi_freio,
    "rota_motor": rota_motor,
    "ang_direct": [random.randint(-30, 30) for _ in range(num_obs)],  # Graus
    "cond_estrada": cond_estrada,
    "carga_truck": carga_truck,
    "consu_combus": consu_combus,
    "cond_clima": [random.choice([0, 1, 2]) for _ in range(num_obs)],  # Ruim, Média, Boa
    "cond_condu": cond_condu
}
df = pd.DataFrame(data)
```

# rede neural

```python
import math
import numpy as np

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def output_activation(x):
    return max(0, min(1, sigmoid(x)))

class NeuralNetwork:
    def __init__(self):
        # Pesos
        self.weights = [0.5] * 102
        # Bias
        self.biases = [0.5] * 13

    def compute_gradients(self, x, h1, h2, y_true, y_pred):
        dw = [0] * len(self.weights)
        db = [0] * len(self.biases)

        error = -(y_true - y_pred)

        for i in range(6):
            dw[78 + i] = error * h2[i]
            db[12] = error

        for i in range(6):
            for j in range(6):
                dw[66 + i*6 + j] = error * self.weights[78 + i] * h2[i] * (1 - h2[i]) * h1[j]
                db[6 + i] = error * self.weights[78 + i] * h2[i] * (1 - h2[i])

        for i in range(6):
            for j in range(10):
                if len(x) > j:
                    dw[i*10 + j] = 0
                    for k in range(6):
                        dw[i*10 + j] += error * self.weights[78 + k] * h2[k] * (1 - h2[k]) * self.weights[66 + k*6 + i] * h1[i] * (1 - h1[i]) * x[j]
                db[i] = 0
                for k in range(6):
                    db[i] += error * self.weights[78 + k] * h2[k] * (1 - h2[k]) * self.weights[66 + k*6 + i] * h1[i] * (1 - h1[i])

        return dw, db

    def feedforward(self, x):
        x = np.array(x).flatten()
        if len(x) < 10 or len(self.weights) < 102:
            print('Length of input vector x:', len(x))
            print('length of weight vector:', len(self.weights))
            return [0]*6, [0]*6, 0

        h1 = [0] * 6
        for i in range(6):
            for j in range(10):
                if len(x) > j:
                    h1[i] += self.weights[i*10 + j] * x[j]
            h1[i] = sigmoid(h1[i] + self.biases[i])

        h2 = [0] * 6
        for i in range(6):
            for j in range(6):
                h2[i] += self.weights[66 + i*6 + j] * h1[j]
            h2[i] = sigmoid(h2[i] + self.biases[6 + i])

        o = 0
        for i in range(6):
            o += self.weights[78 + i] * h2[i]
        o = output_activation(o + self.biases[12])

        return h1, h2, o

    def train(self, X_train, y_train, epochs, learning_rate):
     for epoch in range(epochs):
        for x, y_true in zip(X_train.values, y_train):
            x = np.array(x)

            h1, h2, y_pred = self.feedforward(x)

            dw, db = self.compute_gradients(x, h1, h2, y_true, y_pred)
            self.weights = [w - learning_rate * dw_i for w, dw_i in zip(self.weights, dw)]
            self.biases = [b - learning_rate * db_i for b, db_i in zip(self.biases, db)]

network = NeuralNetwork()
network.train(X_train, y_train, epochs=100, learning_rate=0.1)
```

```python
y_pred = []
for x in X_test.values:
    _, _, o = network.feedforward(x) # avaliação
    y_pred.append(o)

for i, prob in enumerate(y_pred):
    print(f"Probabilidade da amostra {i}: {prob}")
```
